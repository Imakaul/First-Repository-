# compute the following probablities:


\begin{equation}
p(target=0 | (1\leqslant var_0 \leqslant 15)\cap (-10 \leqslant var_1 \leqslant 9.5)\cap (4 \leqslant var_2 \leqslant 18))

\begin{equation}

\begin{equation}
p((1 \leqslant var_0 \leqslant 15)\cap (-10 \leqslant var_1 \leqslant 9.5)\cap (4 \leqslant var_2 \leqslant 18) | target=1)

# lets compute the first one:

\begin{equation}
p(target=0 | (1\leqslant var_0 \leqslant 15)\cap (-10 \leqslant var_1 \leqslant 9.5)\cap (4 \leqslant var_2 \leqslant 18))
\end{equation}


# lets say event (A) = (target = 0 ), & event (b) = (1<= VARo<=15 and -10<=VAR1<=9.5 and 4<=VAR2 <= 18)


# now according to the conditional probablity formula:

\begin{equation}
p(A|B) = \frac{p(A\cap B)}{p(B)}
\end{equation}

# Therefore:

\begin{equation}
p(target=0 | (1\leqslant var_0 \leqslant 15)\cap (-10 \leqslant var_1 \leqslant 9.5)\cap (4 \leqslant var_2 \leqslant 18)) = \frac{p((target=0 \cap (1\leqslant var_0 \leqslant 15)\cap (-10 \leqslant var_1 \leqslant 9.5)\cap (4 \leqslant var_2 \leqslant 18))}{p(1\leqslant var_0 \leqslant 15)\cap (-10 \leqslant var_1 \leqslant 9.5)\cap (4 \leqslant var_2 \leqslant 18)}
\end{equation}
